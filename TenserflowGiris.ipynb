## veriyi test / train olarak ikiye ayırmak

```python
from sklearn.model_selection import train_test_split
```

```python
#train_test_split ctrl+tab ile dökümantasyonuna bak
```

```python
dataFrame
```

```python
#y=wx+b
#y-> label
y=dataFrame["Fiyat"].values #Fiyat değerlerini bir diziye atadık

# x->feature(özellik)
x=dataFrame[["BisikletOzellik1","BisikletOzellik2"]].values

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

## scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential #katman çağırmak için import ettik
from tensorflow.keras.layers import Dense#dense ile katman çağırdık
```

```python
model=Sequential()

model.add(Dense(4,activation="relu"))#5 nöron kullanılacağını belirttik ve aktivasyon fonksiyonu olarak da relu kullanıyoruz
model.add(Dense(4,activation="relu"))
model.add(Dense(4,activation="relu"))

model.add(Dense(1))#çıkış katmanı

model.compile(optimizer="rmsprop",loss="mse")
```

```python
model.fit(x_train,y_train,epochs=250)
```

```python
loss=model.history.history["loss"]
```

```python
sbn.lineplot(x=range(len(loss)),y=loss)
```

```python
trainLoss=model.evaluate(x_train,y_train,verbose=0)
tesstLoss=model.evaluate(x_test,y_test,verbose=0)
```
