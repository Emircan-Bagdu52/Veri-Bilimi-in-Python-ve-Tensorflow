```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**2
```

```python
numpyDizisi2
```

```python
benimFigur, benimEksen = plt.subplots()
benimEksen.plot(numpyDizisi1,numpyDizisi2,"#0000FF",alpha=0.5)#hex code ile istediğimiz renk tonunu seçebiliriz grafik için
#alpha ile saydamlık katabiliyoruz
benimEksen.plot(numpyDizisi2,numpyDizisi1,"b")
```

```python
# Yeni bir figür (çizim alanı) ve eksen (alt çizim alanı) oluşturun.
yeniFigur, yeniEksen = plt.subplots()
# Kırmızı renkte (r) bir çizgi çizin, numpyDizisi1'in değerlerini kullanarak numpyDizisi1 + 2'nin grafiğini çiz.
yeniEksen.plot(numpyDizisi1,numpyDizisi1+2,"r",linewidth=8.0)#linewidth ile grafik çizgisinin kalınlığını belirleyebiliyoruz
yeniEksen.plot(numpyDizisi1,numpyDizisi1+3,"b")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+4,"black",linestyle="--")#linestyle="--" çizginin nasıl olacağını ayarlayabiliyoruz
yeniEksen.plot(numpyDizisi1,numpyDizisi1+5,"black",linestyle="-.")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+6,"black",linestyle=":")
yeniEksen.plot(numpyDizisi1,numpyDizisi1+7,"black",linestyle="--",marker="o",markersize=8,markerfacecolor="red")
#marker ile grafiğimizdeki noktaları işaretleyebiliriz boyutu markersize ile ayarlayabiliriz ve içi boş olan dairelerin içini de renklendirebiliriz

```
