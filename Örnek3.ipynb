```python
y=dataFrame["Type"].values
x= dataFrame.drop("Type",axis=1).values
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
#sınıftan bir obje oluşturuyoruz
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

```python
x_train.shape
```

```python
model=Sequential()

model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1,activation="sigmoid"))


model.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1)
```

```python
#lossları sözlük formatında görme
model.history.history
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()#overfitting(aşırı öğrenme) olayı gerçekleşti
```

```python
model=Sequential()

model.add(Dense(30,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(15,activation="relu"))
model.add(Dense(1,activation="sigmoid"))


model.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25)
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
modelKaybi=pd.DataFrame(model.history.history)
```

```python
modelKaybi.plot()#early stopping ile aşırı öğrenmenin önüne geçildi.
```

### dropout ile verdiğimiz orana göre input,layeri modelimizden rast gele atacak 

```python
model=Sequential()

model.add(Dense(30,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(15,activation="relu"))
model.add(Dropout(0.6))
model.add(Dense(1,activation="sigmoid"))


model.compile(optimizer="adam",loss="binary_crossentropy")
```

```python
model.fit(x=x_train,y=y_train,epochs=700,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
```

```python
kayipDF=pd.DataFrame(model.history.history)
```

```python
kayipDF.plot()
```
