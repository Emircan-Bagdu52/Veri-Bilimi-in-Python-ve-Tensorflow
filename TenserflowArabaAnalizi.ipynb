```python
# nümerik bir değer olmadığı için transmission değerinden kurtuluyoruz
dataFrame=dataFrame.drop("transmission",axis=1)
```

```python
y=dataFrame["price"].values #y değerine veri setindeki fiyatları atadık
x=dataFrame.drop("price",axis=1).values #x değerine de geri kalan sütunları atadık
```

```python
y
```

```python
x
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler=MinMaxScaler()
```

```python
x_train=scaler.fit_transform(x_train)
```

```python
x_test=scaler.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #modeli oluşturuyoruz
from tensorflow.keras.layers import Dense #Katmanları oluşturuyoruz
```

```python
x_train.shape # 5 farklı özellik olduğu için minimum 5 katman kullanmamız gerekir.
```

```python
model=Sequential()

model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))

model.add(Dense(1))

model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250,epochs=300)
```

```python
kayipVerisi=pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()#öğrenmek için gayet normal bir tablo
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
tahminDizisi=model.predict(x_test)
```

```python
tahminDizisi
```

```python
mean_absolute_error(y_test,tahminDizisi)#test ile tahmin arasındaki fark 3384 ortalama
```
