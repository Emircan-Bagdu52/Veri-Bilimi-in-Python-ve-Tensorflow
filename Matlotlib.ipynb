```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline #grafikleri plt.show yazmadan görmemizi sağlar
```

```python
yasListesi=[10,20,30,30,30,40,50,60,70,75]
```

```python
kiloListesi=[20,60,80,85,86,87,70,90,95,90]
```

```python
numpyYasListesi=np.array(yasListesi)
```

```python
numpyKiloListesi=np.array(kiloListesi)
```

```python
numpyYasListesi
```

```python
numpyKiloListesi
```

```python
plt.plot(numpyYasListesi,numpyKiloListesi,"r") #r red anlamına gelir ve grafiğin hangi renk olacağını seçmek için kullanılır.
#ilk olarak yas yazdığımız için o x eksenine ikinci yazdığımız da y eksenine gitti.
plt.xlabel("YAS") #eksenlere başlık verme
plt.ylabel("Kilo")
plt.title("Kilonun Yaşa Göre Değişim Grafiği") # tabloya başlık verme
#plt.show() #%matplotlib inline yazmasaydık her seferinde plt.show yazmak zorunda kalacaktık. 
```

```python
numpyDizisi1=np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2=numpyDizisi1**3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
#g-- yaparsak kesikli noktalar ile grafiği çizer
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"b*-")
#g*- yaparsak yıldızlar ile noktaları işaretler ve grafiği çizer * yerine + koyarsak + ile noktaları gösterir:
```

```python
plt.subplot(1,2,1)#1 sıra olacak ,2kolon olacak 1.grafik
plt.plot(numpyDizisi1,numpyDizisi2,"b--")
plt.subplot(1,2,2)#1 sıra olacak ,2kolon olacak 2.grafik
plt.plot(numpyDizisi2,numpyDizisi1,"g+-")
```

```python
benimFigure=plt.figure() #boş bir figür her tablo bir figürdür.
figureAxes=benimFigure.add_axes([0.2,0.2,0.9,0.9])#sol eksenin başlangıcı, alt eksenin başlangıcı, genişliği ,yüksekliği
#shift tab e basarak nasıl kullanıldığını görebilirsiniz
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X EKSENİ VERİ İSMİ ")
figureAxes.set_ylabel("Y EKSENİ VERİ İSMİ ")
figureAxes.set_title("GRAFİK BAŞLIĞI")
```

```python
figure2=plt.figure()

eksen1=figure2.add_axes([0.1,0.1,0.8,0.8])
eksen2=figure2.add_axes([0.2,0.4,0.4,0.4])

eksen1.plot(numpyDizisi1,numpyDizisi2,"g")
eksen1.set_xlabel("X Ekseni")
eksen1.set_ylabel("Y Ekseni")
eksen1.set_title("Ana Grafik Başlığı")

eksen2.plot(numpyDizisi2,numpyDizisi1,"b")
eksen1.set_xlabel("X Ekseni Küçük")
eksen1.set_ylabel("Y Ekseni Küçük")
eksen1.set_title("Ana Grafik Başlığı Küçük")
```

```python
yeniFigur=plt.figure(dpi=100)#figsize=(5,5) ile de boyutu ayarlanabiliyor dpi=100 ile görüntü kalitesi ayarlanabiliyor 
yeniEksen=yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1**2,label="numpyDizisi **2")#label ile grafik üzerinde çizgilerin adlarını gösterdik.
yeniEksen.plot(numpyDizisi1,numpyDizisi1**3,label="numpyDizisi **3")
yeniEksen.legend(loc=1)#loc ile label da yazdıklarımızın nerede duracağını ayarlıyoruz
```

```python
yeniFigur.savefig("benimfigur1.png",dpi=500)
```
